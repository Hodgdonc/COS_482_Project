```python
fig, (ax0, ax1, ax2) = plt.subplots(3,1, sharex=True)
plt.subplots_adjust(hspace = 0.6)

ax0.plot(merged.loc[merged['Horror'] == 1].groupby([merged.release_year]).agg('count')['Horror'])
ax0.title.set_text("Movies Released per Year in no NaN Dataset")
ax0.set_ylabel("Number of Movies")
ax0.tick_params('x',reset=True)
ax0.sharey(ax1)

ax1.plot(merged2.loc[merged2['Horror'] == 1].groupby([merged2.release_year]).agg('count')['Horror'])
ax1.title.set_text("Movies Released per Year in Complete Dataset")
ax1.set_ylabel("Number of Movies")
ax1.tick_params('x',reset=True)

ax2.plot(merged2.loc[merged2['Horror'] == 1].groupby([merged2.release_year]).agg('count')['Horror'] / merged2.groupby([merged2.release_year]).agg('count')['Horror'])
ax2.title.set_text("Frequency of Genre each Year")
ax2.set_ylabel("Frequency of Genre")
ax2.set_ylim([0, 1])

plt.xlabel("Year")
fig.suptitle("Horror Movies")


```
