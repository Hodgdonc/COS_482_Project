```python
import pandas as pd
df2 = pd.read_csv("imdb_movies_no_selenium.csv",index_col=0)
df2
```

```python
df2_no_NaN = df2.dropna()
df2_no_NaN
```

```python
df2_no_NaN['release_year'] = df2_no_NaN['release_year'].str[1:-1]
df2_no_NaN['gross_US_CA'] = df2_no_NaN['gross_US_CA'].str[1:-1]
df2_no_NaN['length'] = df2_no_NaN['length'].str[:-3]
df2_no_NaN
```

```python
df2_no_NaN.to_csv("imdb_movies_cleaned.csv")
```
