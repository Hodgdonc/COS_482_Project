```python
merged = pd.merge(left=df, left_index=True,right=df3, right_index=True,how='inner')
```

```python
merged
```

```python
merged['release_year'] = merged['release_year'].str.extract(pat='(\d+)',expand=False)
merged
```

```python

horror = merged.loc[merged['Horror'] == 1].groupby([merged.release_year]).agg('count')
horror.plot.line(title='Number of Horror Movies Each Year', ylabel = "Number of Horror Movies", legend=False, color='red', figsize=(5,2), ylim=(0,120))
```
